#Data Cleaning

##Handling Missing Values

```python
df_chicago_crimes.isnull().sum()
```

```python
#Dropping the missing values & checking result
df_chicago_crimes.dropna(inplace=True)
df_chicago_crimes.isnull().sum()
```

```python
df_chicago_crimes.shape
```

After removing null values we lost 4,884 rows which is a small percentage.

##Handling Duplicate Values

```python
#Checking for duplicate records
df_chicago_crimes.duplicated().sum()
```

There are no duplicated records

##Handling Outliers

```python
df_chicago_crimes.boxplot()
```

```python
#Skewness
#Positive values on the ID column indicate that the tail of the data is left-skewed. The skew of the year is negligible hence symmetrical/non-skewed.
print('Skewness of ID values: ' +str(df_chicago_crimes['ID'].skew()))
print('Skewness of Year: ' +str(df_chicago_crimes['Year'].skew()))
```

```python
#Checking Kurtosis
print('Kurtosis of ID Values: ' +str(df_chicago_crimes['ID'].kurt()))
```
